```python
# Define base path
base = "/Volumes/main/default/Inventory_Sales_Data"

# Read all 4 raw CSVs
df_sales = spark.read.format("csv").option("header","true").option("inferSchema","true")\
    .load(f"{base}/Sales_Shipment_Data.csv")

df_inventory = spark.read.format("csv").option("header","true").option("inferSchema","true")\
    .load(f"{base}/Inventory_Stock_Data.csv")

df_supply = spark.read.format("csv").option("header","true").option("inferSchema","true")\
    .load(f"{base}/supply_chain_data.csv")

df_store = spark.read.format("csv").option("header","true").option("inferSchema","true")\
    .load(f"{base}/retail_store_inventory.csv")

print("Sales rows:", df_sales.count())
print("Inventory rows:", df_inventory.count())
print("Supply chain rows:", df_supply.count())
print("Store inventory rows:", df_store.count())
```

```python
# Save to Bronze layer as Delta tables with column mapping enabled
df_sales.write.format("delta").mode("overwrite") \
    .option("delta.columnMapping.mode", "name") \
    .saveAsTable("bronze_sales")

df_inventory.write.format("delta").mode("overwrite") \
    .option("delta.columnMapping.mode", "name") \
    .saveAsTable("bronze_inventory")

df_supply.write.format("delta").mode("overwrite") \
    .option("delta.columnMapping.mode", "name") \
    .saveAsTable("bronze_supply_chain")

df_store.write.format("delta").mode("overwrite") \
    .option("delta.columnMapping.mode", "name") \
    .saveAsTable("bronze_store_inventory")

print("✅ All Bronze tables created!")
```

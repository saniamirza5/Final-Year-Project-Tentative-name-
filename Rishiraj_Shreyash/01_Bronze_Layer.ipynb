```python
base = "/Volumes/main/default/Inventory_Sales_Data"

df_sales = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(f"{base}/Sales_Shipment_Data.csv")

df_inventory = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(f"{base}/Inventory_Stock_Data.csv")

df_supply = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(f"{base}/supply_chain_data.csv")

df_store = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(f"{base}/retail_store_inventory.csv")

print("✅ All files loaded!")
print("Sales rows:", df_sales.count())
print("Inventory rows:", df_inventory.count())
print("Supply chain rows:", df_supply.count())
print("Store inventory rows:", df_store.count())
```

```python
print("=== SALES COLUMNS ===")
df_sales.printSchema()

print("=== INVENTORY COLUMNS ===")
df_inventory.printSchema()

print("=== SUPPLY CHAIN COLUMNS ===")
df_supply.printSchema()

print("=== STORE INVENTORY COLUMNS ===")
df_store.printSchema()
```

```python
df_sales.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("bronze_sales")

df_inventory.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("bronze_inventory")

df_supply.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("bronze_supply_chain")

df_store.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("bronze_store_inventory")

print("✅ Bronze layer complete! All 4 Delta tables saved.")
```
